```python
#Erosion without using the built-in function for morphology

r,c = img.shape
erode = np.zeros((r,c),dtype=int)
for i in range(1,(r-1)):
  for j in range(1,(c-1)):
    temp = [img[i-1,j-1],img[i-1,j],img[i-1,j+1],
            img[i,j-1],img[i,j],img[i,j+1],
            img[i+1,j-1],img[i+1,j],img[i+1,j+1]]

    erode[i,j]=min(temp)

plt.figure(figsize=(10,10))

plt.subplot(121)

plt.imshow(img,cmap='gray')

plt.title('Original')

plt.subplot(122)

plt.imshow(erode,cmap='gray')

plt.title('Erode image')
```

# Dilation 

In the dilation operation, if for ANY pixel that is true (1/shaded) in the structuring element, the corresponding pixel in the image underneath is also true (1/shaded) , then the output pixel is set to true. Otherwise, the output pixel is set to false.

image.png

```python
#Read the image and find the size of the image
img=cv2.imread('Image_for_dilation.png',0)
r,c = img.shape
dilate = np.zeros((r,c),dtype=int)
for i in range(1,(r-1)):
  for j in range(1,(c-1)):
    temp = [img[i-1,j-1],img[i-1,j],img[i-1,j+1],
            img[i,j-1],img[i,j],img[i,j+1],
            img[i+1,j-1],img[i+1,j],img[i+1,j+1]]

    dilate[i,j]=max(temp)

plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img,cmap='gray')
plt.title('Original')

plt.subplot(122)
plt.imshow(dilate,cmap='gray')
plt.title('Dilated image')
```
