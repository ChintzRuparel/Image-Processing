To convert a color (RGB) image to Grayscale

This example converts an image with RGB channels into an image with a single grayscale channel.

Y = 0.2125 R + 0.7154 G + 0.0721 

These weights are used by CRT phosphors as they better represent human perception of red, green and blue than equal weights

```python
from skimage import io
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
img=io.imread('cat.png')
I=rgb2gray(img)
plt.figure(figsize=(12,12)) #increase the figure size
#original image
plt.subplot(121)
io.imshow(img)
plt.title('Original image')
#gray image
plt.subplot(122)
io.imshow(I)
plt.title('gray scale image')
```

While converting a color to gray image you can loose some information 
Ishihara plates are used to check color blindness

```python
from skimage import io
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
img=io.imread('ishihara.jpg')
I=rgb2gray(img)
plt.subplot(121)
io.imshow(img)
plt.title('Original image')
plt.subplot(122)
io.imshow(I)
plt.title('gray scale image')
```

Image Blending

```python
from skimage.transform import resize
I1=io.imread('dhoni.jpg')/255
I2=io.imread('ViratKohli.jpg')/255
I2=resize(I2,(I1.shape[0],I1.shape[1],3))
I=I1+I2
io.imshow(I)
plt.title('blended image')



```

```python
import numpy as np
I1=io.imread('dhoni.jpg')/255
I2=io.imread('ViratKohli.jpg')/255
I2=resize(I2,(I1.shape[0],I1.shape[1],3))
i=1
plt.figure(figsize=(15,15))
for alpha in np.linspace(0,1,5):
  I=alpha*I1+(1-alpha)*I2
  plt.subplot(1,5,i)
  plt.imshow(I)
  i=i+1
```
